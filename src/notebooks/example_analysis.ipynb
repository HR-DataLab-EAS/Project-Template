# Example Analysis Notebook

This notebook demonstrates how to load, process, and analyze data using the project modules.

## Setup

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from src.data import load_raw_data, basic_cleaning, save_processed_data
from src.analysis import calculate_summary_statistics, get_correlation_matrix

# Set styling
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

## Load Data

```python
# Load raw data
df = load_raw_data('sample_data.csv')

# Display first few rows
print(f"Data shape: {df.shape}")
print(f"\nFirst few rows:")
print(df.head())

print(f"\nData types:")
print(df.dtypes)

print(f"\nMissing values:")
print(df.isnull().sum())
```

## Data Cleaning

```python
# Clean data
df_clean = basic_cleaning(df)

print(f"Cleaned data shape: {df_clean.shape}")
print(df_clean.head())
```

## Analysis

```python
# Calculate summary statistics
stats = calculate_summary_statistics(df_clean)

print("Summary Statistics:")
for col, stat in stats.items():
    print(f"\n{col}:")
    for key, value in stat.items():
        print(f"  {key}: {value:.2f}")
```

```python
# Get correlation matrix
corr = get_correlation_matrix(df_clean)

print("Correlation Matrix:")
print(corr)

# Visualize correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
# Age distribution
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
df_clean['age'].hist(bins=10, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Income distribution
plt.subplot(1, 2, 2)
df_clean['income'].hist(bins=10, edgecolor='black', color='orange')
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Save Processed Data

```python
# Save processed data
save_processed_data(df_clean, 'processed_sample.csv')
print("Processed data saved to Data/processed/processed_sample.csv")
```
